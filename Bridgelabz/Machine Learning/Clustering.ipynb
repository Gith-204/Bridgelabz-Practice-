```python
from kmodes.kmodes import KModes
import pandas as pd
import numpy as np
np.random.seed(42)
data = {
    'Gender':    ['Male', 'Female', 'Female', 'Male', 'Male', 'Female',
                  'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Education': ['Graduate', 'PostGrad', 'Graduate', 'Graduate', 'PostGrad', 'PostGrad',
                  'Graduate', 'Graduate', 'PostGrad', 'Graduate', 'PostGrad', 'Graduate'],
    'City':      ['Chennai', 'Mumbai', 'Chennai', 'Delhi', 'Mumbai', 'Delhi',
                  'Chennai', 'Mumbai', 'Chennai', 'Delhi', 'Mumbai', 'Chennai'],
    'Product':   ['Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics', 'Clothing',
                  'Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics', 'Clothing'],
    'Subscribed':['Yes', 'No', 'Yes', 'No', 'Yes', 'No',
                  'Yes', 'No', 'Yes', 'No', 'Yes', 'No']
}
df = pd.DataFrame(data)
print("Customer Dataset (all categorical):\n", df)

costs = []
for k in range(2,6):
  km = KModes(n_clusters=k, init='Huang', n_init=5, verbose=0)
  km.fit(df)
  costs.append(km.cost_)
  print(f"K={k} Cost: {km.cost_:.2f}")

km = KModes(n_clusters=3, init='Huang', n_init=5, verbose=0)
km.fit(df)
clusters = km.predict(df)
df['Cluster'] = clusters

print("\nCluster Assignments: ")
print(df)

print("\nCluster Modes (most frequent category per feature per cluster):")
for i, mode in enumerate(km.cluster_centroids_):
  print(f"\n  Cluster {i}: ")
  for feat, val in zip(df.columns[::-1], mode):
    print(f"   {feat}: {val}")
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
x,y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)
model = KMeans(n_clusters=4, init='k-means++', n_init=10, max_iter=300, random_state=42)
model.fit(x)
labels = model.labels_
centroids = model.cluster_centers_
print(f"Inertia (within-cluster sum of squares): {model.inertia_}")
print(f"Cluster Centers:\n{centroids}")
new_data = np.array([[0, 0], [4, 4]])
predictions = model.predict(new_data)
print(f"Predictions for new data: {predictions}")
inertias = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x)
    inertias.append(kmeans.inertia_)
```
